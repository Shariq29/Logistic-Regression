## Q NO 1: What is Simple Linear Regression (SLR)? Explain its purpose.

## ANS:

**Simple Linear Regression (SLR)** is a statistical method used to model the relationship between two variables: a dependent variable (often denoted as \( y \)) and a single independent variable (denoted as \( x \)). The relationship is represented by a linear equation of the form:

[ y = beta_0 + beta_1 * x + epsilon ]

where:
- ( y ): The dependent variable (the outcome or response variable you’re trying to predict).
- ( x ): The independent variable (the predictor or input variable).
- ( beta_0 ): The intercept (the value of \( y \) when \( x = 0 \)).
- ( beta_1 ): The slope (the change in \( y \) for a one-unit increase in \( x )).
- ( epsilon ): The error term (captures random variation or noise not explained by the model).

### Purpose of Simple Linear Regression
SLR is used to:
1. **Understand Relationships**: Determine whether and how strongly two variables are linearly related. For example, how does study time (\( x \)) affect test scores (\( y \))?
2. **Predict Outcomes**: Use the independent variable to predict the dependent variable. For instance, predicting someone’s house price based on its size.
3. **Quantify Impact**: Estimate the magnitude of the effect of the independent variable on the dependent variable (via the slope \( \beta_1 \)).
4. **Simplify Analysis**: Provide a straightforward way to model and interpret relationships when only one predictor is involved.

### Key Assumptions
For SLR to be valid, certain assumptions must hold:
- The relationship between \( x \) and \( y \) is linear.
- Errors (\( \epsilon \)) are independent, normally distributed, and have constant variance (homoscedasticity).
- The independent variable \( x \) is measured without error.

### Example
If you want to predict a person’s weight (\( y \)) based on their height (\( x \)), SLR can estimate the linear relationship and provide a formula to predict weight from height.


















## Q NO 2: What are the key assumptions of Simple Linear Regression?

## ANS:

The key assumptions of Simple Linear Regression (SLR) are:

1. **Linearity**: The relationship between the independent variable (\( x \)) and the dependent variable (\( y \)) is linear. This means the change in \( y \) is proportional to the change in \( x \).

2. **Independence**: Observations are independent of each other. The value of the dependent variable for one observation does not influence the value of another.

3. **Normality**: The residuals (errors, \( \epsilon \)) are normally distributed. This is particularly important for hypothesis testing and constructing confidence intervals.

4. **Homoscedasticity**: The variance of the residuals is constant across all values of the independent variable (\( x \)). This means the spread of residuals should be roughly the same at all levels of \( x \).

5. **No Measurement Error in \( x \)**: The independent variable (\( x \)) is measured without error, meaning it is a controlled or precisely measured variable.

6. **No Multicollinearity (for SLR specifically, not applicable)**: Since SLR involves only one independent variable, multicollinearity is not a concern, but this assumption becomes relevant in multiple linear regression.

Violations of these assumptions may lead to biased or unreliable results, so they should be checked (e.g., via residual plots or statistical tests) before interpreting the model.





## Q NO 3: Write the mathematical equation for a simple linear regression model and explain each term.

## ANS:

The mathematical equation for a **Simple Linear Regression (SLR)** model is:

[ y = beta_0 + beta_1 * x + epsilon]

### Explanation of Each Term:
1. **\( y \)**: The **dependent variable** (also called the response or outcome variable). This is the variable you aim to predict or explain. For example, in a model predicting house prices based on size, \( y \) would be the house price.

2. **\( beta_0 \)**: The **intercept**. This represents the expected value of \( y \) when the independent variable \( x \) is zero. It’s the point where the regression line crosses the y-axis. For instance, in the house price example, \( beta_0 \) might represent a baseline price when the house size is zero (though this may not always be practically meaningful).

3. **\( beta_1 \)**: The **slope** or **coefficient** of the independent variable. It quantifies the change in \( y \) for a one-unit increase in \( x \). A positive \( \beta_1 \) indicates that as \( x \) increases, \( y \) increases, while a negative \( beta_1 \) indicates the opposite. For example, if \( beta_1 = 2 \), a one-unit increase in house size (e.g., 1 square foot) increases the price by 2 units (e.g., $2).

4. **\( x \)**: The **independent variable** (also called the predictor or explanatory variable). This is the variable used to predict \( y \). In the house price example, \( x \) could be the size of the house in square feet.

5. **\( epsilon \)**: The **error term** (or residual). This represents the portion of \( y \) that the model cannot explain. It captures random variation, measurement errors, or the influence of other unmodeled factors. The error term is assumed to have a mean of zero, be normally distributed, and have constant variance (homoscedasticity) for the model to be valid.

### Purpose of the Equation:
This equation models the linear relationship between \( x \) and \( y \), allowing you to predict \( y \) for a given \( x \) and understand how changes in \( x \) affect \( y \). The parameters \( \beta_0 \) and \( \beta_1 \) are estimated from data using methods like least squares, which minimizes the sum of squared residuals.

## Q NO 5: What is the method of least squares in linear regression?

## ANS:

The **method of least squares** is a statistical technique used in **Simple Linear Regression (SLR)** to find the best-fitting line that models the relationship between the independent variable (\( x \)) and the dependent variable (\( y \)). It achieves this by minimizing the sum of the squared differences (residuals) between the observed values of \( y \) and the values predicted by the linear model.

### Mathematical Explanation
In SLR, the model is given by:

[ y = beta_0 + beta_1 * x + epsilon ]

where:
- \( y \): Observed dependent variable.
- \( beta_0 \): Intercept.
- \( beta_1 \): Slope.
- \( x \): Independent variable.
- \( epsilon \): Error term.

The predicted value of \( y \) for a given \( x \) is:

[ hat{y} = beta_0 + beta_1 * x ]

The **residual** for each data point is the difference between the observed \( y_i \) and the predicted \( hat{y}_i \):

\[ e_i = y_i - hat{y}_i \]

The method of least squares seeks to find the values of \( \beta_0 \) and \( \beta_1 \) that minimize the **sum of squared residuals** (SSR):

\[ text{SSR} = sum_{i=1}^n e_i^2 = sum_{i=1}^n (y_i - (beta_0 + beta_1 x_i))^2 \]

By minimizing SSR, the method ensures the regression line is as close as possible to all data points.

### How It Works
1. **Objective**: Find \( \beta_0 \) and \( \beta_1 \) that make the sum of squared residuals as small as possible.
2. **Solution**: The optimal values of \( \beta_0 \) and \( \beta_1 \) are derived using calculus (by taking partial derivatives of SSR with respect to \( \beta_0 \) and \( \beta_1 \), setting them to zero, and solving). The resulting formulas are:

   - Slope (\( beta_1 \)):

     \[ beta_1 = frac{sum_{i=1}^n (x_i - bar{x})(y_i - bar{y})}{sum_{i=1}^n (x_i - bar{x})^2} \]

     where \( bar{x} \) and \( bar{y} \) are the means of \( x \) and \( y \), respectively.

   - Intercept (\( beta_0 \)):

     \[ beta_0 = bar{y} - beta_1 bar{x} \]

3. **Result**: These formulas yield the line \( hat{y} = beta_0 + beta_1 x \) that best fits the data by minimizing the total squared error.

### Why Square the Residuals?
- Squaring ensures residuals are positive, preventing positive and negative errors from canceling each other out.
- Squaring penalizes larger errors more heavily, prioritizing a closer fit for outliers.
- The method results in a unique solution for \( beta_0 \) and \( beta_1 \).

### Real-World Context
In the example of predicting exam scores (\( y \)) based on study hours (\( x \)), the method of least squares would use observed data (e.g., hours studied and corresponding scores) to calculate \( \beta_0 \) and \( beta_1 \). The resulting regression line (e.g., \( hat{y} = 50 + 5x \)) minimizes the sum of squared differences between actual scores and the predicted scores on the line.

### Key Points
- The method assumes the relationship is linear and errors are normally distributed with constant variance.
- It’s computationally efficient and widely used in statistical software to fit regression models.
- The resulting line provides the best linear approximation of the data, enabling predictions and insights into the relationship between \( x \) and \( y \).

This method is foundational for SLR and extends to multiple linear regression for models with more than one predictor.












## Q NO 6: What is Logistic Regression? How does it differ from Linear Regression?

## ANS:

### What is Logistic Regression?

**Logistic Regression** is a statistical method used for binary classification problems, where the goal is to predict the probability that an observation belongs to one of two categories (e.g., 0 or 1, yes or no, pass or fail). Unlike Simple Linear Regression (SLR), which predicts a continuous outcome, logistic regression predicts the probability of an event occurring, which is then mapped to a binary outcome using a threshold (e.g., 0.5).

The model uses the **logistic function** (sigmoid function) to constrain the output to a range between 0 and 1, representing probabilities. The mathematical form of logistic regression is:

\[ P(y=1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x)}} \]

where:
- \( P(y=1|x) \): Probability that the outcome is 1 given the input \( x \).
- \( \beta_0 \): Intercept.
- \( \beta_1 \): Coefficient for the independent variable \( x \).
- \( e \): Base of the natural logarithm.
- The term \( \beta_0 + \beta_1 x \) is the linear combination of inputs, similar to linear regression.

For multiple predictors, the model extends to:

\[ P(y=1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n)}} \]

The model is fitted by maximizing the **likelihood function** (or minimizing the log-loss) rather than using least squares, as in linear regression.

**Purpose**:
- Predict the probability of a binary outcome (e.g., whether a patient has a disease based on medical test results).
- Classify observations into one of two categories (e.g., spam vs. not spam for emails).
- Understand the impact of predictors on the likelihood of an event.

**Example**: Predicting whether a student passes (1) or fails (0) an exam based on study hours (\( x \)). The output is the probability of passing, which can be thresholded (e.g., >0.5 = pass).

---

### Differences Between Logistic Regression and Linear Regression

| **Aspect**                     | **Linear Regression**                                                                 | **Logistic Regression**                                                              |
|--------------------------------|--------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------|
| **Purpose**                    | Predicts a continuous dependent variable (e.g., house price, exam score).             | Predicts the probability of a binary outcome (e.g., pass/fail, yes/no).             |
| **Output**                     | Continuous values (unbounded, can be any real number).                                | Probabilities (bounded between 0 and 1) or binary classification after thresholding. |
| **Model Equation**             | \( y = \beta_0 + \beta_1 x + \epsilon \)                                             | \( P(y=1|x) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x)}} \)                           |
| **Function Used**              | Linear function (straight line).                                                      | Logistic (sigmoid) function to map to probabilities.                                |
| **Fitting Method**             | Minimizes the sum of squared residuals (least squares).                               | Maximizes the likelihood function (or minimizes log-loss).                          |
| **Dependent Variable**         | Continuous (e.g., weight, temperature).                                               | Binary or categorical (e.g., 0/1, true/false).                                      |
| **Assumptions**                | - Linearity between \( x \) and \( y \).<br>- Normality of residuals.<br>- Homoscedasticity.<br>- Independent observations. | - Linear relationship between predictors and the log-odds.<br>- Independent observations.<br>- No multicollinearity (for multiple predictors). |
| **Error Term**                 | Assumes errors (\( \epsilon \)) are normally distributed with constant variance.       | No explicit error term; errors are modeled via the logistic function.                |
| **Interpretation of Coefficients** | \( \beta_1 \): Change in \( y \) per unit change in \( x \).                          | \( \beta_1 \): Change in the log-odds of the outcome per unit change in \( x \).     |
| **Example Use Case**           | Predicting house price based on square footage.                                       | Predicting whether a customer will buy a product (yes/no) based on age.             |

---

### Key Conceptual Differences
1. **Nature of the Problem**:
   - Linear regression is used for **regression tasks** where the outcome is continuous (e.g., predicting someone’s weight).
   - Logistic regression is used for **classification tasks** where the outcome is categorical (e.g., predicting whether a tumor is malignant or benign).

2. **Output Transformation**:
   - Linear regression directly outputs a continuous value.
   - Logistic regression transforms the linear combination (\( \beta_0 + \beta_1 x \)) using the sigmoid function to produce a probability, which is then used for classification.

3. **Interpretation**:
   - In linear regression, the slope (\( \beta_1 \)) directly indicates the change in the dependent variable.
   - In logistic regression, the slope affects the **log-odds** (\( \ln\left(\frac{P(y=1)}{P(y=0)}\right) \)), making interpretation less intuitive but related to the likelihood of the outcome.

4. **Model Fitting**:
   - Linear regression uses the least squares method to minimize errors.
   - Logistic regression uses maximum likelihood estimation, as the binary outcome doesn’t fit the least squares framework.

---

### Real-World Example for Comparison
- **Linear Regression**: Predicting a student’s exam score (e.g., 85 out of 100) based on hours studied. The model might be \( \text{Score} = 50 + 5 \cdot \text{Hours} \).
- **Logistic Regression**: Predicting whether a student passes (1) or fails (0) based on hours studied. The model might output \( P(\text{Pass}) = \frac{1}{1 + e^{-(-2 + 0.5 \cdot \text{Hours})}} \), and a threshold (e.g., 0.5) determines the classification.





## Q NO 7: Name and briefly describe three common evaluation metrics for regression models.

## ANS:

Here are **three common evaluation metrics for regression models**, each with a brief description:

1. **Mean Absolute Error (MAE)**: This metric calculates the average of the absolute differences between predicted and actual values. It represents the average error magnitude in the same units as the target variable, making it easy to interpret. A lower MAE indicates better model accuracy.

- MAE measures the average absolute difference between predicted values ($\hat{y}_i$) and actual values ($y_i$), calculated as $\text{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|$. It provides a straightforward measure of average error magnitude in the same units as the target variable, making it intuitive. MAE treats all errors equally, regardless of size.

2. **Mean Squared Error (MSE)**: MSE computes the average of the squared differences between predicted and actual values. By squaring the errors, it penalizes larger errors more heavily, emphasizing significant deviations. A lower MSE indicates better model performance, but it’s in squared units of the target variable.

- MSE calculates the average of the squared differences between predicted and actual values, given by $\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2$. Squaring the errors amplifies larger discrepancies, making MSE sensitive to outliers. It’s useful for penalizing significant errors but is less interpretable due to squared units.

3. **Root Mean Squared Error (RMSE)**: RMSE is the square root of MSE, providing a measure of the standard deviation of prediction errors. It’s expressed in the same units as the target variable, making it more interpretable than MSE. A lower RMSE indicates a better-fitting model.

- RMSE is the square root of MSE, computed as $\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2}$. It measures the standard deviation of prediction errors in the same units as the target variable, making it easier to interpret than MSE. RMSE balances sensitivity to large errors with interpretability.


## Q NO 8: What is the purpose of the R-squared metric in regression analysis?

## ANS:

The **R-squared** metric, also known as the **coefficient of determination**, measures the proportion of the variance in the dependent variable (\( y \)) that is explained by the independent variable(s) in a regression model. Its purpose in regression analysis is to:

1. **Assess Model Fit**: R-squared indicates how well the regression model fits the data. It ranges from 0 to 1, where:
   - \( R^2 = 0 \): The model explains none of the variance in the dependent variable.
   - \( R^2 = 1 \): The model explains all the variance in the dependent variable.
   - For example, an \( R^2 = 0.85 \) means 85% of the variability in \( y \) is explained by the model.

2. **Evaluate Predictive Power**: It provides insight into how well the independent variable(s) predict the dependent variable. A higher R-squared suggests a stronger linear relationship between the variables in Simple Linear Regression (SLR).

3. **Compare Models**: R-squared helps compare different models for the same dataset. A model with a higher R-squared is generally better at explaining the data, though it should be considered alongside other metrics (e.g., RMSE, MAE) to avoid overfitting concerns.

### Calculation
R-squared is calculated as:

[ R^2 = 1 - SSR / SST ]

where:
- **SSR** (Sum of Squared Residuals): \( \sum (y_i - \hat{y}_i)^2 \), the sum of squared differences between actual (\( y_i \)) and predicted (\( \hat{y}_i \)) values.
- **SST** (Total Sum of Squares): \( \sum (y_i - \bar{y})^2 \), the total variance in the dependent variable, where \( \bar{y} \) is the mean of \( y \).

### Limitations
- **Context Matters**: A high R-squared doesn’t guarantee a good model; it may indicate overfitting, especially with complex models.
- **Not for Non-Linear Models**: R-squared is less reliable for non-linear relationships.
- **Doesn’t Imply Causation**: A high R-squared only shows association, not that \( x \) causes \( y \).

In Simple Linear Regression, R-squared is particularly useful for assessing how well a single predictor explains the outcome. For example, in predicting exam scores from study hours, an \( R^2 = 0.7 \) means 70% of the variability in scores is explained by study hours.

```python
# Q NO 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

# ANS:


# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate sample data with y as 1D array (standard for single-target regression)
np.random.seed(42)  # For reproducibility
x = np.random.rand(100) * 10  # 1D array for independent variable (e.g., study hours)
X = x.reshape(-1, 1)     # Reshape to 2D array (n_samples, n_features) as required by scikit-learn
y = 3 + 2 * x + np.random.randn(100) * 2  # 1D dependent variable (e.g., exam scores)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Extract the slope (coefficient) and intercept as scalars
slope = model.coef_[0]      # coef_ is array([value]) for 1D y → extract [0]
intercept = model.intercept_  # intercept_ is a scalar float for 1D y

# Print the slope and intercept
print(f"Slope (β₁): {slope:.2f}")
print(f"Intercept (β₀): {intercept:.2f}")

# Optional: Visualize the data and regression line
plt.scatter(X[:, 0], y, color='blue', label='Data points')
plt.plot(X[:, 0], model.predict(X), color='red', label='Regression line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

## Q NO 10: How do you interpret the coefficients in a simple linear regression model?

## ANS:

In a **Simple Linear Regression (SLR)** model, the relationship between the independent variable (\( x \)) and the dependent variable (\( y \)) is expressed by the equation:

\[ y = beta_0 + beta_1 * x + epsilon \]

The coefficients \( \beta_0 \) and \( \beta_1 \) have specific interpretations:

1. **Intercept (\( beta_0 \))**:
   - **Definition**: The intercept represents the predicted value of the dependent variable (\( y \)) when the independent variable (\( x \)) is equal to zero.
   - **Interpretation**: It is the baseline value of \( y \) when \( x = 0 \), assuming the model is valid at \( x = 0 \) and the intercept is meaningful in the context of the data. For example, in a model predicting exam scores (\( y \)) based on study hours (\( x \)), if \( \beta_0 = 50 \), it suggests a student who studies zero hours is expected to score 50 points (though this may not always be practically meaningful if \( x = 0 \) is outside the data range).

2. **Slope (\( beta_1 \))**:
   - **Definition**: The slope represents the change in the dependent variable (\( y \)) for a one-unit increase in the independent variable (\( x \)).
   - **Interpretation**: It quantifies the strength and direction of the relationship between \( x \) and \( y \).
     - If \( beta_1 > 0 \), an increase in \( x \) is associated with an increase in \( y \).
     - If \( beta_1 < 0 \), an increase in \( x \) is associated with a decrease in \( y \).
     - For example, if \( beta_1 = 2 \) in the exam score model, each additional hour of study (\( x \)) increases the predicted exam score (\( y \)) by 2 points.

### Example
Suppose you fit an SLR model to predict house prices (\( y \), in thousands of dollars) based on house size (\( x \), in square feet), and the model is:

\[ \hat{y} = 100 + 0.15x \]

- **Intercept (\( \beta_0 = 100 \))**: When the house size is 0 square feet, the predicted price is $100,000. This may not be realistic (as houses don’t have zero size), but it serves as the baseline.
- **Slope (\( \beta_1 = 0.15 \))**

 For every additional square foot, the predicted house price increases by $150 (since 0.15 × 1000 = 150).

  So, a 100-square-foot increase in size raises the predicted price by $15,000.

### Key Considerations
- **Context Matters**: The interpretation of coefficients depends on the units of \( x \) and \( y \). Always specify what a “one-unit increase” in \( x \) means (e.g., 1 hour, 1 square foot).
- **Causation vs. Correlation**: \( \beta_1 \) shows the association between \( x \) and \( y \), not necessarily causation. Other factors may influence \( y \).
- **Range of Data**: Interpretations are most reliable within the range of observed \( x \) values. Extrapolating (e.g., at \( x = 0 \)) can be misleading if it’s outside the data range.
- **Statistical Significance**: Check if \( \beta_1 \) is statistically significant (e.g., using p-values) to ensure the relationship isn’t due to random chance.
